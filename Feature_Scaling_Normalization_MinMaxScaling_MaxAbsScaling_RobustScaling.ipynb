```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malic_acid']
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic_acid'])
```

```python
color_dict={1:'red',3:'green',2:'blue'}
sns.scatterplot(df['Alcohol'],df['Malic_acid'],hue=df['Class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=0.2,random_state=0)



x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
```

```python
# fit the scaler to the train set, it will learn the parameters
Scaler.fit(x_train)
```

```python
# transform train and test sets
X_train_scaled = Scaler.transform(x_train)
X_test_scaled = Scaler.transform(x_test)
```

```python
#converting numpy in to dataframe
X_train_scaled = pd.DataFrame(X_train_scaled,columns=x_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(X_test_scaled.describe(),1)
```

```python
fig,(ax1,ax2)= plt.subplots(ncols =2 ,figsize =(12,5))

ax1.scatter(x_train['Alcohol'],x_train['Malic_acid'],c=y_train)
ax1.set_title("Before Scalling")
ax2.scatter(X_train_scaled['Alcohol'],X_train_scaled['Malic_acid'],c=y_train)
ax2.set_title("After Scalling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax=ax1)
sns.kdeplot(x_train['Malic_acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(X_train_scaled['Malic_acid'], ax=ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax=ax1)


# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
```
